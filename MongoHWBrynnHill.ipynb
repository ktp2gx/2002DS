```python
from pymongo import MongoClient
import pprint
```

```python
connection_string = "mongodb+srv://db2002ds:Hattie13DS_@cluster0.2posm.mongodb.net/test?retryWrites=true&w=majority"
client = MongoClient(connection_string)
db = client['sample_mflix']
collection = db['movies']
```

```python
#1.1 Write a query to find the first movie that has the genre "Action".
action_movies = collection.find({"genres": "Action"}).limit(1)
for movie in action_movies:
    pprint.pprint(movie)
```

```python
#1.2 Write a query to find all movies released after the year 2000 (Return the first 5 results)
greater_2000_movies = collection.find({ "year": {"$gt": 2000 } }).limit(5)
for movie in greater_2000_movies:
    pprint.pprint(movie)
```

```python
#1.3 Write a query to find all movies where the IMDb rating is greater than 8.5 (Return the
#first 5 results).
high_rated_movies = collection.find({"imdb.rating": {"$gt": 8.5}}).limit(5)
for movie in high_rated_movies:
    pprint.pprint(movie)
```

```python
#1.4 Write a query to find all movies where the genre contains both "Action" and "Adventure"
action_adventure_movies = collection.find({"genres": {"$all": ["Action", "Adventure"]}}).limit(5)
for movie in action_adventure_movies:
    pprint.pprint(movie)
```

```python
#2.1 Write a query to find all movies where the genre is "Comedy" and sort them by IMDb
# rating in descending order (Return the first 5 results)
comedy_movies = collection.find({"genres": "Comedy"}).sort("imdb.rating", -1).limit(5)
for movie in comedy_movies:
    pprint.pprint(movie)
```

```python
#2.2 Write a query to find all movies where the genre is "Drama" and sort them by release year
# in ascending order (Return the first 5 results).
drama_movies = collection.find({"genres": "Drama"}).sort("year", 1).limit(5)
for movie in drama_movies:
    pprint.pprint(movie)
```

```python
#3.1 Write an aggregation pipeline that calculates the average IMDb rating for each genre
# (Return the top 5 genres).
avg_rating_by_genre = collection.aggregate([
    {"$unwind": "$genres"},
    {"$group": {"_id": "$genres", "average_rating": {"$avg": "$imdb.rating"}}},
    {"$sort": {"average_rating": -1}},
    {"$limit": 5}
])
for genre in avg_rating_by_genre:
    pprint.pprint(genre)
```

```python
#3.2 Write an aggregation pipeline to find the top 5 directors by the average IMDb rating of
# their movies.
top_5_directors = collection.aggregate([
    {"$group": {"_id": "$directors", "average_rating": {"$avg": "$imdb.rating"}}},
    {"$sort": {"average_rating": -1}},
    {"$limit": 5}
])
for director in top_5_directors:
    pprint.pprint(director)
```

```python
#3.3 Write an aggregation pipeline to calculate the total number of movies released in each
# year (Sort the results by the year).
movies_by_year = collection.aggregate([
    {"$group": {"_id": "$year", "total_movies": {"$sum": 1}}},
    {"$sort": {"_id": 1}}
])
for year in movies_by_year:
    pprint.pprint(year)
```

```python
#4.1 Write a query to update the IMDb rating of a movie with the title "The Godfather" to 9.5.
collection.update_one({"title": "The Godfather"}, {"$set": {"imdb.rating": 9.5}})

#4.2 Write a query to update all movies where the genre is "Horror" and set their IMDb rating
#to 6.0 if it is currently null
collection.update_many({"genres": "Horror", "imdb.rating":{"$exists":False}}, {"$set": {"imdb.rating": 6.0}})

#4.3 Write a query to delete all movies that were released before the year 1950.
collection.delete_many({"year": {"$lt": 1950}})
```

```python
#5.1 Ensure the title field is indexed for text search in MongoDB and write a query to search
#for movies that contain the word "love" in their title
collection.create_index([("title")])
love_movies = collection.find({"$text": {"$search": "love"}})
for movie in love_movies:
    pprint.pprint(movie)
```

```python
#5.2 Write a text search query to find movies where the word "war" appears in the title or plot,
# sorted by IMDb rating (Return only the top 5 results).
collection.create_index([("title"), ("plot")])
war_movies = collection.find({"$text": {"$search": "war"}}).sort("imdb.rating", -1).limit(5)
for movie in war_movies:
    pprint.pprint(movie)
```

```python
#6.1 Write a query to find all movies where the genre is "Action" and the IMDb rating is
# greater than 8 (Sort the results by release year in descending order).
action_movies = collection.find({"genres": "Action", "imdb.rating": {"$gt": 8}}).sort("year", -1)
for movie in action_movies:
    pprint.pprint(movie)
```

```python
# 6.2 Write a query to find all movies where the directors contain "Christopher Nolan" and the
# IMDb rating is greater than 8 (Return only the top 3 movies sorted by rating)
action_high_rated_movies = collection.find({"directors": "Christopher Nolan", "imdb.rating": {"$gt": 8}}).sort("imdb.rating", -1).limit(3)
for movie in action_high_rated_movies:
    pprint.pprint(movie)
```
