```python
import pandas as pd

acc_players_df = pd.read_csv('/content/acc_players-2324F.csv')

acc_players_df.head()
```

```python
acc_players_df.columns = acc_players_df.iloc[0]
acc_players_df = acc_players_df.drop(0).reset_index(drop=True)
acc_players_df.head()
```

```python
#2 Calculate the total points scored by all players combined

acc_total_PTS = acc_players_df['PTS'].sum()
print(acc_total_PTS)
```

```python
#2 Find the player who has played the most minutes (MP)
most_minutes_player = acc_players_df.loc[acc_players_df['MP'].idxmax()]['Player']
print(most_minutes_player)
```

```python
#2 Identify the top 5 players in terms of total rebounds (TRB)
acc_players_df['TRB'] = pd.to_numeric(acc_players_df['TRB'])

top_5_rebounds = acc_players_df.nlargest(5, 'TRB')['Player']
print(top_5_rebounds)
```

```python
#3 Create a new DataFrame containing only players who played more than 500 minutes.
acc_players_df['MP'] = pd.to_numeric(acc_players_df['MP'])
playing_players_df = acc_players_df[acc_players_df['MP'] > 500]
print(playing_players_df)
```

```python
#3 player with the highest total assists (AST).
highest_assists = playing_players_df.loc[playing_players_df['AST'].idxmax()]['Player']
print(highest_assists)
```

```python
#3 Top 3 Assist Leads in the League
playing_players_df['AST'] = pd.to_numeric(playing_players_df['AST'])
top_3_assists = playing_players_df.sort_values(by='AST', ascending=False).head(3)['Player']
print(top_3_assists)
```

```python
#3 Top 3 Shot Blockers
playing_players_df['STL'] = pd.to_numeric(playing_players_df['STL'])
top_3_shot_blockers = playing_players_df.sort_values(by='STL', ascending=False).head(3)['Player']
print(top_3_shot_blockers)
```

```python
#4 Group the players by School and calculate the total points scored by each school.
acc_players_df['PTS'] = pd.to_numeric(acc_players_df['PTS'])
school_points = acc_players_df.groupby('School')['PTS'].sum()
print(school_points)
```

```python
#4 Group the players by School and calculate the total assists (AST) for each team
school_assists = acc_players_df.groupby('School')['AST'].sum()
print(school_assists)
```

```python
#4 Sort the schools by total points scored and display the top 3 schools
top_3_schools = school_points.nlargest(3)
print(top_3_schools)
```
